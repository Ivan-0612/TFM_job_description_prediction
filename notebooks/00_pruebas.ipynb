## Modelo LLM en local

Probe los siguientes modelos: qwen3:8b-q8_0, qwen2.5:14b y qwen2.5:14b-instruct

Voy a probar los modelos de arriba con ollama en local. He escogido estos modelos ya lo soporta la vram de mi tarjeta gráfica, 12 gb y son lo suficientemente potente como para utilizarlo tanto para extraer información necesaria de la descripción . Además son un entrenado con datos estructurados y código, es más fiable a la hora de consistencia de formato, que es justamente lo que necesito.

Para la parte de generación de ofertas de trabajo el mejor modelo es gemma 3 12b, sigo pudiendo ejecutarlo en mi tarjeta gráfica y es ideal para generar texto.

```python
import ollama

prompt = """ Eres un experto en análisis de datos y tu tarea es ayudarme a obtener datos de ofertas de empleo en España.
Quiero que me proporciones una lista de 20 sectores laborales y 20 puestos de trabajo """

response = ollama.chat(
    model = 'qwen2.5:7b-instruct',
    messages=[{'role':'user', 'content': prompt}]
)
```

```python
print(response['message']['content'])
```

Existen 3 tipos de roles, a cada rol le puedes enviar un contenido:
- user: la tarea que quieres que el modelo procese
- system: define las reglas del modelo, como se debe comportar, su personalidad y restricciones.
- assistant: se utiliza para mostrar al modelo ejemplos de respuesta. Tanto de contenido como de formato.

```python
sectores_generales = [
    "Agricultura",
    "Ganadería",
    "Pesca",
    "Minería",
    "Industria",
    "Energía",
    "Construcción",
    "Comercio",
    "Turismo",
    "Hostelería",
    "Transporte",
    "Logística",
    "Finanzas",
    "Seguros",
    "Tecnología",
    "Comunicaciones",
    "Educación",
    "Sanidad",
    "Inmobiliario",
    "Administración",
    "Cultura",
    "Deportes",
    "Servicios"
]


oferta_ejemplo = """
TÍTULO DEL PUESTO: Reponedor/a de Frutas y Verduras
FUNCIONES DEL PUESTO:
- Gestión de producto: Ayuda al montaje de sección, manipulación, mantenimiento y conservación del producto.
- Gestión comercial y atención al cliente: Potenciar la decisión de compra del cliente mediante asesoramiento técnico sobre el producto, la calidad y orientación personalizada.
- Gestión de las relaciones dentro del equipo: Apoyo directo al Jefe de Sección.
- Gestión en materia de calidad alimentaria y seguridad y salud: Control de caducidades, limpieza de la sección siguiendo normativa legal y uso adecuado de los EPIS.

COMPETENCIAS:
Identidad Corporativa, Orientación al cliente, Trabajo en equipo, Organización, rigor y calidad, Orientación a resultados y Competencia Digital.

REQUISITOS:
- Formación: Ciclo formativo de Grado Medio o ESO.
- Formación complementaria: Conocimiento de paquete Office.
- Experiencia: Mínima de 6 meses en el puesto requerido.

OTROS REQUERIMIENTOS:
Manejo de herramientas y utillajes fundamentales de la sección. Conocimientos técnicos sobre el producto, así como habilidad y destreza manual para trabajar con el mismo.
"""

prompt = f"""
    Eres un experto en reclutamiento IT. 
    Analiza el texto de la oferta y extrae la información en JSON, en español, traduce si es necesario.
    Si no encuentras un dato, pon null.
    Recibiras ofertas de empleo en texto plano, y tu tarea es extraer la información relevante y estructurarla en formato JSON.
    JSON:
    - años_experiencia: int (en años, si no se especifica, pon null, si es menos de un año, pon 0)
    - experiencia: intern, junior, senior o null. Para determinar esto, si los años de experiencia son 0 o se especifica en la oferta, será intern. Si son entre 1 y 3, será junior. Si son más de 3, será senior. Sino encuentras años de experiencia, pon null, al menos que la oferta especifique claramente que es para un puesto de intern, junior o senior (o nombres similares).
    - modalidad: 'Remoto', 'Presencial' o 'Hibrido'
    - habilidades: un unico string con las habilidades, competencias, funciones o responsabilidades requeridas.
    - formación académica: 'Ninguna', 'Educación Secundaria Obligatoria (ESO)', 'Formación Profesional de Grado Medio', 'Formación Profesional de Grado Superior', 'Grado Universitario' o 'Postgrado', según lo que se especifique en la oferta, en caso de encontrar varia formación requerida poner la de nivel inferior. Si no se especifica, pon null.
    - beneficios: lista de strings
    - salario: string
    - sector: uno de los siguientes sectores en base a la oferta: {', '.join(sectores_generales)}
    - tipo de empleo: 'Jornada completa', 'Media jornada', 'Temporal', 'Contrato por obra', 'Prácticas', 'Voluntariado' o 'Otro' (sino se especifica: null)
    - ciudad: solo de las 17 comunidades autónomas y 2 ciudades autónomas posibles, sino especifica null
    """

response = ollama.chat(
    model = 'qwen2.5:14b-instruct',
    messages=[{'role':'system', 'content': prompt},
              {'role':'user', 'content': oferta_ejemplo}],
    format='json',
    options={'temperature': 0.0}
)

print(response['message']['content'])
```

- qwen3:8b-q8_0 tardo 1,46 minutos
- qwen2.5:14b tardo 9,2 segundos
- qwen2.5:14b-instruct tardo 8,3 segundos. Este último está más orientado al seguimiento de instrucciones
- qwen2.5:7b-instruct tardo 6,3 segundos. Tiene menos parámetros y puede funcionar peor, probaré con este y el de 14b

Parámetros que se pueden configurar:
- temperature: que tan creativo o arriesgado es el modelo. 
    - Entre 0 y 0.3: modelo determinista, escoge siempre la opción más probable (lo que yo necesito en la parte de extracción, cuanto más bajo mejor).
    - Entre 0.7 y 1.2: modelo más creativo, escoge palabras menos probables.
- top_p: para configurar el % de palabras posibles que el modelo mira. Con 0.1, el modelo solo considera el 10% de las palabras más probables.
- top_k: lo mismo que top_p, pero de forma absoluta. Con 40, el modelo considera las 40 palabras más probables.
- seed: similar a random_state. Si mantienes la misma seed y temperature, el modelo siempre responderá lo mismo. 
- num_predict: el límite de tokens que el modelo puede generar por respuesta. Puede evitar que el modelo divage y genere demasiado.
- num_gpu: 0 para que use la CPU y -1 para que use la GPU.
