```python
df = pd.DataFrame()
df_glassdoor = pd.DataFrame() #dataframe vacio para ir llenandolo

for f in files:
    df = pd.read_csv(f)
    categoria = f.split('\\')[1].split('_')[0]

    if categoria == 'ciudad':
        df['Localizacion'] = df['Busqueda_Ciudad']
        df['Sector'] = np.nan
    elif categoria == 'puesto':
        df['Sector'] = df['Busqueda_Puesto']
    else:
        df['Sector'] = np.nan

    df_glassdoor = pd.concat([df,df_glassdoor],axis=0)
```
