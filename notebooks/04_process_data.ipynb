Voy ha hacer una limpieza de las columnas con texto (titulo, habilidades y beneficios)

```python
# TITULO
# elimino () y lo que hay dentro (y espacios antes y después)
df_final['Titulo'] = df_final['Titulo'].str.replace(r'\s*\([^)]*\)', '', regex=True).str.strip()
# elimino caracteres especiales
df_final['Titulo'] = df_final['Titulo'].str.replace(r'\\W', '', regex=True)
# elimino espacios extra
df_final['Titulo'] = df_final['Titulo'].str.replace(r'\\s+', ' ', regex=True)

# Beneficios
# elimino caracteres especiales
df_final['beneficios'] = df_final['beneficios'].str.replace(r'\\W', '', regex=True).str.strip()
# elimino espacios extra
df_final['beneficios'] = df_final['beneficios'].str.replace(r'\\s+', ' ', regex=True)

# Habilidades
df_final['habilidades'] = df_final['habilidades'].str.replace(r'\\s+', ' ', regex=True).str.strip()
# no elimino caracteres especiales porque pueden ser parte de las habilidades (C++)
```

```python
df_final.head()
```

```python
df_final.shape
```

Voy a crear una columna de salario medio (media entre salario mínimo y máximo) y una columna de salarios categórica con un rango salarial. Esto para facilitar el EDA.

```python
rangos = ["<15.000","15.000-22.000","22.000-30.000","30.000-40.000","40.000-52.000","52.000-65.000","65.000-80.000","80.000-100.000","100.000-150.000", ">150.000"]
df_final['Salario_medio'] = (df_final['salario_minimo'] + df_final['salario_maximo']) / 2
df_final['Rango_Salarial'] = pd.cut(df_final['Salario_medio'], bins=[0, 15000, 22000, 30000, 40000, 52000, 65000, 80000, 100000, 150000, float('inf')], labels=rangos)
```

Con esto ya no necesito las variables de salario mínimo y salario máximo.

```python
df_final.drop(columns=['salario_minimo', 'salario_maximo'], inplace=True)
```

### Título y Habilidades

Voy a vectorizar las columas de título y habilidades para poder entrenar más adelante modelos con esta información. Voy a utilizar SBERT ya que es capaz de entender el contexto de cada palabra. Utilizó SBERT en lugar de BERT ya que este último tiene la capacidad de entender cada palabra individal de una frase dentro de su contexto mientras que SBERT genera un único vector para cada frase que representa su significado global, lo que facilita comparar miles de filas rápidamente.

```python
# Primero genero una columna única para título y habilidades.
df_final['Titulo_Habilidades'] = df_final['Titulo'] + ' ' + df_final['habilidades']

# Limpiar cadenas
df_final['Titulo_Habilidades'] = df_final['Titulo_Habilidades'].astype(str).str.strip()
df_final['Titulo_Habilidades'] = df_final['Titulo_Habilidades'].str.lower()

# guardo los textos preparados en una lista para generar los embeddings
textos = ["passage: " + texto for texto in df_final['Titulo_Habilidades'].tolist()] 
#passage es necesario para que el modelo de embeddings sepa que es un texto a procesar
```

```python
from sentence_transformers import SentenceTransformer
import torch
import sys
# Comprobación rápida del intérprete y CUDA
print('python executable:', sys.executable)
print('torch:', getattr(torch, '__version__', None), 'built with CUDA', getattr(torch.version, 'cuda', None))
print('cuda_available:', torch.cuda.is_available())
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print('Using device:', device)
# Cargar el modelo SBERT usando detección automática de device
model = SentenceTransformer('intfloat/multilingual-e5-large', device=device)

# Generar los embeddings para los textos
matriz_embeddings = model.encode(
    textos, 
    show_progress_bar=True, 
    batch_size=64 
)
```
